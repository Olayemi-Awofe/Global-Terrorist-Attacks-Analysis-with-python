```python
terror_attack = df.groupby("Country")['Terrorist attacks'].sum()
terror_attack
```

```python
terror_attack.sort_values(ascending = False, inplace = True)
terror_attack
```

```python
terror_attack.sort_values(ascending = False).tail(10)
```

```python
terror_attack.head(10).plot(kind = "bar")
plt.title("Top 10 Numbers of Terrorist attacks per Country")
plt.xlabel("Country")
plt.ylabel("Terrorist Attacks")
plt.xticks(rotation = 90)
plt.show
```

```python
terror_attack.tail(10).plot(kind = "bar")
plt.title("Below 10 Numbers of Terrorist attacks per Country")
plt.xlabel("Country")
plt.ylabel("Terrorist Attacks")
plt.xticks(rotation = 90)
plt.show
```

```python
country_count = df["Country"].value_counts()
country_count
```

```python
country_count.head(10).plot(kind = "bar")
plt.title("Numbers of times country was attack by Terrorist")
plt.xlabel("Country")
plt.ylabel("Numbers of Attacks")
plt.xticks(rotation = 90)
plt.show
```

```python
plt.scatter(df["Terrorism deaths"], df["Terrorist attacks"])
plt.title("Scatter plot of Terrorism death Versus Terrorism Attacks")
plt.xlabel("Terrorism Death")
plt.ylabel("Terrorist Attacks")
plt.show
```

```python
# Time Series Analysis
plt.figure(figsize=(12,6))
sns.lineplot(x= "Year", y = "Terrorist attacks", data = df )
plt.title("Trend of Terrorist Attacks Over the Years")
plt.show
```

```python
# Time Series Analysis
plt.figure(figsize=(12,6))
sns.lineplot(x= "Year", y = "Terrorism deaths", data = df )
plt.title("Trend of Terrorism deaths Over the Years")
plt.show
```

```python
# Attack Methods
attack_methods = df.filter(like= "Attack method").sum()
plt.figure(figsize=(12,6))
attack_methods.plot(kind= "bar")
plt.title("Distribution of Attack Methods")
plt.show()
```

```python
# Age Group Analysis
age_columns = df.filter(like= "Death Age")
plt.figure(figsize=(12,6))
sns.barplot(data = age_columns, errorbar = None)
plt.title("Distribution of Deaths Across Age Groups")
plt.show
```

```python
# Correlation Analysis
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_columns].corr()


plt.figure(figsize = (12, 8))
sns.heatmap(correlation_matrix, annot = True, cmap = "coolwarm", fmt= ".2f")
plt.title("Correlation Matrix")
plt.show
```
